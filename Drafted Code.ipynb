```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Load the cleaned GeoJSON data
gdf = gpd.read_file(r"F:\Merged_Geodata.geojson")

# Step 1: Standardizing the Data
# Select independent variables for clustering
independent_vars = ['HDX_2018-2022', 'INC_2018-2022', 'HTJ_2018-2022', 
                    'NCZ_2018-2022', 'LEQ_2018-2022', 'EDB_2018-2022', 
                    'EDE_2018-2022', 'UNS_2018-2022', 'MCD_2018-2022']

# Extract and standardize the independent variables
data = gdf[independent_vars]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 2: Finding Optimal K for KMeans using Elbow Method (Manual)
inertia = []
silhouette_scores = []
k_values = range(2, 15)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(data_scaled, labels))

# Plot the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, 'bo-', markersize=8)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.savefig('elbow_method_manual.png')
plt.show()

# Plot Silhouette Scores for Different K Values
plt.figure(figsize=(10, 6))
plt.plot(k_values, silhouette_scores, 'go-', markersize=8)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k Values')
plt.savefig('silhouette_scores_manual.png')
plt.show()

# Step 3: KMeans Clustering with Optimal K (Assuming k=5 based on elbow and silhouette scores)
optimal_k = 5  # You may adjust this value based on the elbow plot and silhouette scores
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_labels = kmeans.fit_predict(data_scaled)
gdf['KMeans_Cluster'] = kmeans_labels

# Step 4: Silhouette Score for KMeans
kmeans_silhouette = silhouette_score(data_scaled, kmeans_labels)
print(f"Silhouette Score for KMeans with k={optimal_k}: {kmeans_silhouette}")

# Step 5: DBSCAN Clustering Parameter Tuning
# Testing different values for eps
best_eps = 0.1
best_silhouette = -1
for eps in np.arange(0.1, 2.0, 0.1):
    dbscan = DBSCAN(eps=eps, min_samples=5)
    labels = dbscan.fit_predict(data_scaled)
    # Ignore if only one cluster or all noise
    if len(set(labels)) > 1 and -1 not in labels:
        score = silhouette_score(data_scaled, labels)
        if score > best_silhouette:
            best_silhouette = score
            best_eps = eps

print(f"Optimal eps for DBSCAN: {best_eps}, Silhouette Score: {best_silhouette}")

# Step 6: DBSCAN with Optimal Parameters
dbscan = DBSCAN(eps=best_eps, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_scaled)
gdf['DBSCAN_Cluster'] = dbscan_labels

# Step 7: Agglomerative Clustering
# Trying with the same number of clusters as the optimal K from KMeans
agglo = AgglomerativeClustering(n_clusters=optimal_k)
agglo_labels = agglo.fit_predict(data_scaled)
gdf['Agglomerative_Cluster'] = agglo_labels

# Step 8: Silhouette Score Evaluation
agglo_silhouette = silhouette_score(data_scaled, agglo_labels)
print(f"Silhouette Score for Agglomerative Clustering with k={optimal_k}: {agglo_silhouette}")

# Step 9: Visualization with Adjusted Parameters
# Plot KMeans Clustering Result with Optimal K
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
base = gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
gdf.plot(column='KMeans_Cluster', ax=base, cmap='Blues', legend=True, alpha=0.8,aspect=1)
plt.title(f'KMeans Clustering of Chicago Communities (k={optimal_k})')
plt.axis('off')
plt.savefig('kmeans_clustering_optimized.png')
plt.show()

# Plot DBSCAN Clustering Result with Optimal eps
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
base = gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
gdf[gdf['DBSCAN_Cluster'] != -1].plot(column='DBSCAN_Cluster', ax=base, cmap='viridis', legend=True, alpha=0.8,aspect=1)
plt.title(f'DBSCAN Clustering of Chicago Communities (eps={best_eps})')
plt.axis('off')
plt.savefig('dbscan_clustering_optimized.png')
plt.show()

# Plot Agglomerative Clustering Result with Optimal K
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
base = gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
gdf.plot(column='Agglomerative_Cluster', ax=base, cmap='Reds', legend=True, alpha=0.8,aspect=1)
plt.title(f'Agglomerative Clustering of Chicago Communities (k={optimal_k})')
plt.axis('off')
plt.savefig('agglomerative_clustering_optimized.png')
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import OPTICS
from sklearn.metrics import silhouette_score

# Load the cleaned GeoJSON data
gdf = gpd.read_file(r"F:\Merged_Geodata.geojson")

# Step 1: Standardizing the Data
independent_vars = ['HDX_2018-2022', 'INC_2018-2022', 'HTJ_2018-2022',
                    'NCZ_2018-2022', 'LEQ_2018-2022', 'EDB_2018-2022',
                    'EDE_2018-2022', 'UNS_2018-2022', 'MCD_2018-2022']

data = gdf[independent_vars]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 2: Apply PCA for Dimensionality Reduction
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Step 3: OPTICS Clustering
optics = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.1)
optics_labels = optics.fit_predict(data_pca)
gdf['OPTICS_Cluster'] = optics_labels

# Step 4: Calculate Silhouette Score for OPTICS
# Filtering out noise points (-1 labels)
valid_labels = optics_labels[optics_labels != -1]
if len(set(valid_labels)) > 1:
    optics_silhouette = silhouette_score(data_pca[optics_labels != -1], valid_labels)
    print(f"Silhouette Score for OPTICS: {optics_silhouette}")
else:
    optics_silhouette = "Not enough clusters to calculate silhouette score"
    print(optics_silhouette)

# Step 5: Visualization of OPTICS Clustering with Gradient Brown Color
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
base = gdf.plot(ax=ax, color='lightgrey', edgecolor='black')
gdf[gdf['OPTICS_Cluster'] != -1].plot(column='OPTICS_Cluster', ax=base, cmap='copper', legend=True, alpha=0.8)
plt.title('OPTICS Clustering of Chicago Communities (Gradient Brown)')
plt.axis('off')
plt.savefig('optics_clustering_brown.png')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned GeoJSON data
gdf = gpd.read_file(r"F:\Merged_Geodata.geojson")

# Define dependent variables and independent variables
dependent_var_1 = 'VRDIDR_2018-2022'  # Drug-induced mortality rate
dependent_var_2 = 'CZD_2018-2022'      # Drug abuse crimes (classification)
independent_vars = ['HDX_2018-2022', 'INC_2018-2022', 'HTJ_2018-2022',
                    'NCZ_2018-2022', 'LEQ_2018-2022', 'EDB_2018-2022',
                    'EDE_2018-2022', 'UNS_2018-2022', 'MCD_2018-2022']

# Rename columns for easier understanding and readability
readable_vars = {
    'HDX_2018-2022': 'Hardship Index',
    'INC_2018-2022': 'Median Income',
    'HTJ_2018-2022': 'Crowded Housing',
    'NCZ_2018-2022': 'Non-Citizens',
    'LEQ_2018-2022': 'Limited English',
    'EDB_2018-2022': 'High School Graduation',
    'EDE_2018-2022': 'College Graduation',
    'UNS_2018-2022': 'Uninsured Rate',
    'MCD_2018-2022': 'Medicaid Coverage'
}

data = gdf[independent_vars + [dependent_var_1, dependent_var_2]].dropna()
X = data[independent_vars]
X.rename(columns=readable_vars, inplace=True)

# Step 1: Feature Importance for Drug-Induced Mortality Rate (Random Forest Regressor)
y_vrdidr = data[dependent_var_1]
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X, y_vrdidr)

# Get feature importance for mortality rate
feature_importances_vrdidr = rf_reg.feature_importances_
importance_df_vrdidr = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances_vrdidr})
importance_df_vrdidr = importance_df_vrdidr.sort_values(by='Importance', ascending=False)

# Plot feature importance for mortality rate
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df_vrdidr, color='b')
plt.title('Feature Importance in Predicting Drug-Induced Mortality Rate')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('feature_importance_vrdidr.png')
plt.show()

# Step 2: Feature Importance for Drug Abuse Crimes (Random Forest Classifier)
y_czd = data[dependent_var_2].apply(lambda x: 1 if x > data[dependent_var_2].median() else 0)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X, y_czd)

# Get feature importance for drug abuse crimes
feature_importances_czd = rf_clf.feature_importances_
importance_df_czd = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances_czd})
importance_df_czd = importance_df_czd.sort_values(by='Importance', ascending=False)

# Plot feature importance for drug abuse crimes
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df_czd, color='r')
plt.title('Feature Importance in Predicting Drug Abuse Crimes')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('feature_importance_czd.png')
plt.show()

# Step 3: Combined Analysis
importance_df_vrdidr['Type'] = 'Mortality Rate'
importance_df_czd['Type'] = 'Drug Crimes'
combined_df = pd.concat([importance_df_vrdidr, importance_df_czd])

# Plot combined feature importance (Vertical Bar Chart)
plt.figure(figsize=(12, 8))
sns.barplot(x='Feature', y='Importance', hue='Type', data=combined_df, palette=['blue', 'red'])
plt.title('Comparison of Feature Importance for Mortality Rate and Drug Abuse Crimes')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('combined_feature_importance.png')
plt.show()

# Step 4: Tabulate the Results in a Readable Format
# Create tables for feature importance
vrdidr_table = importance_df_vrdidr[['Feature', 'Importance']].copy()
czd_table = importance_df_czd[['Feature', 'Importance']].copy()

# Save tables as images or display them in a readable format
print("\nFeature Importance for Drug-Induced Mortality Rate:\n")
print(vrdidr_table.to_string(index=False))

print("\nFeature Importance for Drug Abuse Crimes:\n")
print(czd_table.to_string(index=False))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import geopandas as gpd
import pandas as pd

# Load the cleaned GeoJSON data
gdf = gpd.read_file(r"F:\Cleaned_Merged_Geodata.geojson")

# Step 1: Standardizing the Data
independent_vars = ['HDX_2018-2022', 'INC_2018-2022', 'HTJ_2018-2022', 
                    'NCZ_2018-2022', 'LEQ_2018-2022', 'EDB_2018-2022', 
                    'EDE_2018-2022', 'UNS_2018-2022', 'MCD_2018-2022']
data = gdf[independent_vars]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 2: PCA for Dimensionality Reduction
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Step 3: DBSCAN Clustering
# Define eps and min_samples values based on previous testing or trial-and-error
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps value accordingly
dbscan_labels = dbscan.fit_predict(data_pca)

# Add DBSCAN labels to the GeoDataFrame
gdf['DBSCAN_Cluster_PCA'] = dbscan_labels
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity

# Define the resilient communities (resilient)
resilient = gdf[
    (gdf['HDX_2018-2022'] > gdf['HDX_2018-2022'].median()) &
    (gdf['VRDIDR_2018-2022'] < gdf['VRDIDR_2018-2022'].median()) &
    (gdf['EDE_2018-2022'] < gdf['EDE_2018-2022'].median())
]
resilient = resilient.sort_values(by=['HDX_2018-2022'], ascending=False).head(7)

# Define the culturally protected communities (cultural_protection)
cultural_protection = gdf[
    (gdf['NCZ_2018-2022'] > gdf['NCZ_2018-2022'].median()) &
    (gdf['LEQ_2018-2022'] > gdf['LEQ_2018-2022'].median())
]
cultural_protection = cultural_protection.sort_values(by=['NCZ_2018-2022'], ascending=False).head(7)

# Create a combined dataset for Resilient + Cultural Protection Communities
combined_resilient_cultural = pd.concat([resilient, cultural_protection]).drop_duplicates()

# Step 1: Create binary columns for the DBSCAN Yellow Cluster and Combined Communities
# Assuming that the yellow cluster label in DBSCAN clustering is denoted as '1'
yellow_cluster = (gdf['DBSCAN_Cluster_PCA'] == 1).astype(int)

# Create a binary column for the Combined Resilient + Cultural Communities
combined_resilient_cultural_idx = combined_resilient_cultural.index
gdf['Combined_Resilient_Cultural'] = gdf.index.isin(combined_resilient_cultural_idx).astype(int)

# Step 2: Calculate Jaccard Similarity
jaccard_similarity = jaccard_score(yellow_cluster, gdf['Combined_Resilient_Cultural'])
print(f"Jaccard Similarity between Combined Communities and Yellow Cluster: {jaccard_similarity:.2f}")

# Step 3: Calculate Cosine Similarity for Feature Profiles
features = ['HDX_2018-2022', 'INC_2018-2022', 'LEQ_2018-2022', 'HTJ_2018-2022', 'EDB_2018-2022']

# Extract feature means for Yellow Cluster and Combined Communities
yellow_features = gdf[yellow_cluster == 1][features].mean().values.reshape(1, -1)
combined_features = gdf[gdf['Combined_Resilient_Cultural'] == 1][features].mean().values.reshape(1, -1)

# Calculate Cosine Similarity
cosine_sim = cosine_similarity(yellow_features, combined_features)[0][0]
print(f"Cosine Similarity between Yellow Cluster and Combined Communities Feature Profiles: {cosine_sim:.2f}")
```
